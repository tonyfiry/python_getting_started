```python
import matplotlib.pyplot as plt # 匯入matplotlib模組
```

### 繪製長條圖:bar

```python
listx = ['c++','java','python','C#','HTML']
listy = [30,55,65,12,56]
color = ['r','b','g','y','m']
label = ['c++','java','python','C#','HTML']
plt.bar(listx,listy,width=0.8,color=color,label=label)
plt.legend()
plt.rcParams['font.sans-serif'] = 'Microsoft Yahei'
plt.rcParams['axes.unicode_minus'] = False 
plt.title("程式統計",fontsize=18)
plt.xlabel("程式",fontsize=18)
plt.ylabel("人數",fontsize=18)
plt.show()
```

### 筆記(長條圖):
* width:設定每個項目矩形的寬度。
* bottom:設定每個矩形y座標的起始位置，不設定時預測值為0。
* color:設定每個項目矩形的顏色，設定值跟折線圖相同，預測為藍色。
* label:設定每個項目圖例名稱，此屬性必搭配 legend()函數才有效果。

### 繪製橫條圖:barh

```python
listx = ['c++','java','python','C#','HTML']
listy = [45,67,13,89,32]
color = ['r','b','g','y','m']
label = ['c++','java','python','C#','HTML']
plt.barh(listy,listx,height=6,color=color,label=label,left=0)
plt.legend()
plt.ylim(0,100)
plt.rcParams['font.sans-serif'] = 'Microsoft Yahei'
plt.rcParams['axes.unicode_minus'] = False 
plt.title("程式統計",fontsize=18)
plt.xlabel("程式",fontsize=18)
plt.ylabel("人數",fontsize=18)
plt.show()
```

### 筆記(橫條圖):barh:
* y 座標串列:顯示每個項目的名稱串列或是序列串列。
* x 座標串列:顯示每個項目的數值串列。
* height:設定每個項目矩形的高度，不設定預設值是0.8。
* left:設定每個項目矩形x座標的起始位置，不設定時預設值為0。 

### 繪製推疊長條圖

```python
listx = ['c++','java','python','C#','HTML']
list1 = [30,55,65,12,56]
list2 = [12,45,72,6,11]
plt.bar(listx,list1,width=0.5,color='blue',label="男")
plt.bar(listx,list2,width=0.5,color='pink',bottom=list1,label="女")#bottom是設定矩形Y座標的起始位置。
plt.legend()
plt.rcParams['font.sans-serif'] = 'Microsoft Yahei'
plt.rcParams['axes.unicode_minus'] = False 
plt.title("程式統計",fontsize=18)
plt.xlabel("程式",fontsize=18)
plt.ylabel("人數",fontsize=18)
plt.ylim(0,400)
plt.show()
```

### 繪製並列長條圖

```python
import matplotlib.pyplot as plt

width = 0.25
listx = ['c++','java','python','C#','HTML']
listx1 = [x + width/2 for x in range(len(listx))]# 使用串列表達式
listx2 = [x - width/2 for x in range(len(listx))]# 使用串列表達式
listy1 = [45,67,13,89,32]
listy2 = [12,45,25,6,11]
listy1 = [30,62,44,56,52]
listy2 = [45,36,34,36,21]
plt.bar(listx1,listy1,color='blue',label="男",width=width)
plt.bar(listx2,listy2,color='pink',label="女",width=width)#bottom是設定矩形Y座標的起始位置。
plt.legend()
plt.rcParams['font.sans-serif'] = 'Microsoft Yahei'
plt.rcParams['axes.unicode_minus'] = False 
plt.title("程式統計",fontsize=18)
plt.xlabel("程式",fontsize=18)
plt.ylabel("人數",fontsize=18)
plt.xticks(range(len(listx)),labels=listx)#設定座標刻度
plt.show()
```

### 繪製圓餅圖:pie

```python
sizes = [25,30,79,60]
labels = ["歷史","數學","英文","國文"]
explode=[0,0,0.1,0]
colors = ['r','g','y','pink']
plt.rcParams['font.sans-serif'] = 'Microsoft Yahei'
plt.rcParams['axes.unicode_minus'] = False 
plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True,startangle=90,autopct="%2.1f%%",labeldistance=1.1,pctdistance=0.6)
```

### 筆記圓餅圖(pie):
* labels:每一個項目標題組成的串列。
* colors:每一個項目顏色字元組成的字串或是串列。例:colors = ['r','g',b]或是colors = "rgb"
* explode:每一個項目凸出距離數字組成的串列，0表示是正常顯示。
* labeldistance:項目標題與圓心的距離是半徑的多少倍。
* autopct:項目的百分比，語法是「"%2.1f%%」"。
* pctdistance:百分比文字與圓心的距離是半徑的多少倍。
* shadow:布林值:True表示是圖形有陰影，Flase表示是圖形沒有陰影。

### 設定圖表區: figure

```python
def figure1():
    plt.figure(figsize=[10,5],dpi=96,facecolor="w",edgecolor="red",frameon=True)
    listx = ['c++','java','python','C#','HTML']
    listy = [30,55,65,12,56]
    color = ['r','b','g','y','m']
    label = ['c++','java','python','C#','HTML']
    plt.bar(listx,listy,width=0.8,color=color,label=label)
    plt.legend()
    plt.rcParams['font.sans-serif'] = 'Microsoft Yahei'
    plt.rcParams['axes.unicode_minus'] = False 
    plt.title("程式統計",fontsize=18)
    plt.xlabel("程式",fontsize=18)
    plt.ylabel("人數",fontsize=18)
    plt.show()
f1 = figure1()
print(f1)#第一份資料

def figure2():
    plt.figure(figsize=[10,5],dpi=96,facecolor="w",edgecolor="red",frameon=True)
    listx = ['c++','java','python','C#','HTML']
    listy = [20,45,35,62,56]
    color = ['r','b','g','y','m']
    label = ['c++','java','python','C#','HTML']
    plt.bar(listx,listy,width=0.8,color=color,label=label)
    plt.legend()
    plt.rcParams['font.sans-serif'] = 'Microsoft Yahei'
    plt.rcParams['axes.unicode_minus'] = False 
    plt.title("程式統計",fontsize=18)
    plt.xlabel("程式",fontsize=18)
    plt.ylabel("人數",fontsize=18)
    plt.show()
f2 = figure2()
print(f2)#第二份資料

```

### 筆記:(圖表區): figure
* figsize:設定方式為串列，[寬(width)，高(height)]，但單位為英吋表示，例:[6.2,3.2]。
* dpi: 設定解析度，單位為英吋的點數(Dots per inch)。
* facecolor:設定背景顏色，預設值為白色(white)
* edgecolor:設定邊緣顏色，預設值為白色(white)
* frameon:布林值，設定是否有邊框，預設值為True。
* tight_layout:布林值，設定多個圖表間是否邊界預設值為Flase。
